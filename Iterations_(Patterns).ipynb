#Task 10
Hollow Rectangle. Display a rectangle of a given length and width.

Sample Input: 4, 5

Sample Output:

10p.png

Hint 1: Print the character space ( ) in the middle.

Hint 2: You can re-use your solution to PROBLEM 2 and use if condition to selectively print the first and the last star of each line and all-stars of the first and last line.

```python

length=int(input("Please enter the length: "))
width=int(input("Please enter the width: "))
for i in range(1,length+1):
  for j in range(1,width+1):
    if i==1 or i==length:
      print("*",end=" ")
    else:
      if j==1 or j==width:
        print("*",end=" ")
      else:
        print(" ",end=" ")
  print()
```

#Task 17
Rhombus, Just draw the image of the above triangle once. And then, the opposite, once.

```python

row_no = int(input("Please enter the number: "))
for i in range(1,row_no+1):
  for j in range(i,row_no+1):
    print(" ",end=" ")
  for j in range(1,2*i):
    if j==1 or j == 2*i-1 :
      print(j,end=" ")
    else:
      print(" ",end=" ")
  print()
for k in range(1,row_no):
  for l in range(1,k+2):
    print(" ",end=" ")
  for l in range(1,2*(row_no-1)):
    if l == 2*l-1 or l == 2*(row_no-k)-1:
      print(l,end=" ")
    else:
      print(" ",end=" ")
  print()
```

#Task 18
Palindrome.

Palindrome: A palindrome is a word, number, phrase, or other sequence of characters that reads the same backward as forward, such as madam or racecar, 1234321, 62826.

Sample Input: 5

Sample Output:

123454321

```python

num = int(input("Please enter the number: "))
for j in range(1,num+1):
    print(j,end=" ")
for k in range(j-1,0,-1):
    print(k,end=" ")
```

#Task 19
Palindromic Triangle

Sample Input: 5

Sample Output:

19p.png


```python

row_no = int(input("Please enter the number: "))
for i in range(1,row_no+1):
  for j in range(1,row_no-i+1):
    print("#",end=" ")
  for k in range(1,i+1):
    print(k,end=" ")
  for l in range(i-1,0,-1):
    print(l,end=" ")
  for m in range(i,row_no):
    print("#",end=" ")
  print()
```
